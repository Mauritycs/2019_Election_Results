```python
# import data and sort the values by number of people of who voted
counties_summary = pd.read_csv('../Data/counties_summary.csv').sort_values(by=['voted'], ascending=False, inplace=False)
```

```python
counties_summary.reset_index(drop=True)
```

```python
# create lists for bar plot 
```

```python
trace1 = go.Bar(
    x=counties_summary.county.tolist(),
    y=counties_summary.biram.tolist(),
    name='Biram',
    marker=dict(
        color='green',
      
    )
)
trace2 = go.Bar(
    x=counties_summary.county.tolist(),
    y=counties_summary.ghazouani.tolist(),
    name='Ghazouani'
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='stack',
    title = "this is the title"
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')
```
